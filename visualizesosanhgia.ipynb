## Vẽ biểu đồ tần suất của các trang thương mại điện tử

```python
amount_ecommercesite = df['priority_store'].value_counts()
```

```python
plt.subplots(figsize=(20,10))
plt.bar(amount_ecommercesite.index, amount_ecommercesite.values)
plt.xlabel('E-commerce site')
plt.ylabel('Erequency')
plt.xticks(rotation=45)
plt.show()
```
